```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# OpenCV 버전을 확인합니다
print("OpenCV version:")
print(cv2.__version__)
 
# 이미지를 읽어 옵니다. 
img = cv2.imread("/content/gdrive/MyDrive/Basic_Vision_Skills/rear_garden.PNG")
print("width: {} pixels".format(img.shape[1]))
print("height: {} pixels".format(img.shape[0]))
print("channels: {}".format(img.shape[2]))

# 이미지의 높이와 넓이를 가져옵니다.
(height, width) = img.shape[:2]
# 이미지의 중심을 구합니다.
center = (width // 2, height // 2)
 
# 이미지를 Display 합니다.
cv2_imshow(img)

# 이미지 컬러를 BGR to GRAY 로 변환합니다.(GRAY Filter 적용 효과)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

# 이미지 컬러를 BGR to HSV 로 변환합니다.(HSV Filter 적용 효과)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)

# 이미지 컬러를 BGR to LAB 로 변환합니다.(LAB Filter 적용 효과)
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
cv2_imshow(lab)
```
